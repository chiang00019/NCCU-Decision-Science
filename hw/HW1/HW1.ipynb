```python
import numpy as np

def simulate_dice(num_simulations):
    # 公平骰子的選擇機率均等
    fair_dice = [1, 2, 3, 4, 5, 6]

    # 不公平骰子的選擇機率，2的機率是0.25，其他的機率均等
    unfair_dice = [2, 1, 3, 4, 5, 6]
    unfair_probs = [0.25, 0.15, 0.15, 0.15, 0.15, 0.15]  # 設置不公平骰子的機率

    # 儲存最小值的結果
    min_values = []

    # 模擬擲骰子 num_simulations 次
    for _ in range(num_simulations):
        # 擲兩顆公平的骰子
        fair_rolls = np.random.choice(fair_dice, 2, replace=True)

        # 擲一顆不公平的骰子
        unfair_roll = np.random.choice(unfair_dice, p=unfair_probs)

        # 取三顆骰子中的最小值
        min_value = min(fair_rolls[0], fair_rolls[1], unfair_roll)
        min_values.append(min_value)

    # 計算最小值的期望
    expected_value = np.mean(min_values)
    return expected_value

# 模擬 10000 次擲骰子，並計算最小值的期望
expected_min_value = simulate_dice(10000)
print("三顆骰子中最小值的期望值是: " + str(expected_min_value))
```

```python
"""
Q4(30%). You are a manager of SAP, and you just hire a salesperson from Oracle. Suppose 
the salesperson has to visit 20 potential clients this month. Before he/she makes these visits, 
you think high skill or low skill is equally likely. If there is high skill, then the probability of 
making a sale is 2/3 in each of the 20 visits. If there is low skill, then the probability of 
making a sale is 1/3 in each of the 20 visits. Use np.random.choice( ) and whatever functions 
needed in Python to write a simulation program for the case described above. Simulate the 
scenario for 1,000 runs and answer the following questions based on simulation results. 
(a) Suppose you have a policy of promoting the salesperson if there are at least 9 sales in 
these 20 visits. What is the probability this salesperson will be promoted this month? 
(b) What is the conditional probability that this person will be promoted if there is high skill? 
(c) What is the conditional probability that there is high skill given that the salesperson is 
promoted under this policy? 
(d) Make a table showing how the three probabilities in (a), (b), and (c) would change, if the 
policy instead promotes the salesperson when there are at least n sales in the 20 visits for 
n in 1, 2, 3,…, 19, 20. 
"""
import numpy as np

# 固定隨機種子，以便重現結果
np.random.seed(5566)

def question_a(num_simulations, n=9):
    # 紀錄大於等於 n 的次數
    result = 0 
    
    # 重複 num_simulations 次
    for _ in range(num_simulations): 
        # 紀錄成功的次數
        num_success = 0
        
        # 一半的機率是高階，另一半是低階
        skill = np.random.choice(["high", "low"], replace=True)  # 修正此行
        
        # 會有 20 次機會拜訪
        for _ in range(20): 
            # 如果是高階，成功的機率是 2/3，反之是 1/3
            if skill == "high":
                prob = [2/3, 1/3]
            else:
                prob = [1/3, 2/3] 
            
            # 根據隨機選擇是否成功
            success = np.random.choice([1, 0], replace=True, p=prob)
            
            # 如果成功，紀錄下來
            if success == 1:
                num_success += 1
        
        # 如果成功的次數大於等於 n
        if num_success >= n:
            result += 1
            
    # 回傳大於等於 n 的機率
    return result / num_simulations

def question_b(num_simulations, n=9):
    # 紀錄大於等於 n 的次數
    result = 0 
    
    # 重複 num_simulations 次
    for _ in range(num_simulations): 
        # 紀錄成功的次數
        num_success = 0
        
        # 高階情況下進行 20 次拜訪
        for _ in range(20): 
            success = np.random.choice([1, 0], replace=True, p=[2/3, 1/3])
            
            # 如果成功，紀錄下來
            if success == 1:
                num_success += 1
        
        # 如果成功的次數大於等於 n
        if num_success >= n:
            result += 1
            
    # 回傳大於等於 n 的機率
    return result / num_simulations

def question_c(num_simulations, n=9):
    # 紀錄大於等於 n 的次數
    result = 0 
    num_high = 0
    
    # 重複 num_simulations 次
    for _ in range(num_simulations): 
        # 紀錄成功的次數
        num_success = 0
        
        # 一半的機率是高階，另一半是低階
        skill = np.random.choice(["high", "low"], replace=True)  # 修正此行
        
        # 會有 20 次機會拜訪
        for _ in range(20): 
            # 如果是高階，成功的機率是 2/3，反之是 1/3
            if skill == "high":
                prob = [2/3, 1/3]
            else:
                prob = [1/3, 2/3] 
            
            # 根據隨機選擇是否成功
            success = np.random.choice([1, 0], replace=True, p=prob)
            
            # 如果成功，紀錄下來
            if success == 1:
                num_success += 1
        
        # 如果成功的次數大於等於 n
        if num_success >= n:
            if skill == "high":
                num_high += 1   
            result += 1
    
    # 回傳 P(high | promoted)，result可能為 0
    return num_high / result if result != 0 else 0

if __name__ == "__main__":
    # 模擬執行的次數
    simulations = 10000
    print(f"Q1（一般情況下會升遷的機率為）: {question_a(simulations):.4f}") 
    print(f"Q2（高階情況下會升遷的機率為）: {question_b(simulations):.4f}")
    print(f"Q3（升遷的人是高階的機率為）: {question_c(simulations):.4f}")
    print("\nQ4（不同的升遷條件）:")
    
    # 打印表頭
    print(f"{'n':>2} | {'Q1':>10} | {'Q2':>10} | {'Q3':>10}")
    print("-" * 40)
    
    # 打印每一行的數據
    for i in range(1, 21):
        q1 = question_a(simulations, n=i)
        q2 = question_b(simulations, n=i)
        q3 = question_c(simulations, n=i)
        print(f"{i:>2} | {q1:10.4f} | {q2:10.4f} | {q3:10.4f}")
```

## 法一：按照題目機率

```python
"""
Q5(15%) Suppose you go to the college of commerce library (商圖) on Monday evening and 
would like to borrow a book. You are told that the book has been checked out the previous 
Thursday. Assume no one else is waiting for the book. The library staff tells you that 
borrowers return books after 4, 5, 6, or 7 days, with probabilities of 0.1, 0.2, 0.3, 0.4. Note 
that the library is open 7 days a week. 
Hint: The book was checked out on previous Thursday and not available on Monday evening. 
So, the borrower already held the book for 4 days (i.e., Fri, Sat, Sun, and Mon). 
As before, 50% of students return their books to a “foreign” library (社圖/總圖), resulting in 
an extra 2-day delay before the book arrives back to the home library. You decide to check 
the status of the book every evening. What is the probability you will need to wait until 
Wednesday evening to get the book (i.e., the borrow returns the book after 6 days)?   
Write a Monte-Carlo simulation program to compute the probability (Hint: the probability is 
conditional should be close to 4/19).  

解題思路:

直到星期三才能拿到書，可以分成兩種情況:
1. 如果還書的日期是星期一，且還書地點是 foreign library，則可以在星期三拿到書，因為需要等待兩天。
2. 如果還書的日期是星期三，且還書地點是 commercial library，則可以在星期三拿到書。

所以我們模擬這兩種情況，並計算成功的次數，最後計算成功的次數除以總共模擬的次數即可得到機率。
"""
import numpy as np

# 固定隨機種子，以便重現結果
np.random.seed(5566)

if __name__ in "__main__":
    # 模擬執行的次數
    s = 1000000
    # 紀錄可以成功拿到書的次數
    count = 0
    for _ in range(s):
        
        # 隨機選擇還書的地點，CL = commercial library, FL = foreign library
        location = np.random.choice(["CL", "FL"], replace=True)
        
        # 隨機選擇還書的日期
        date_to_return = np.random.choice(
            ["Monday", "Tuesday", "Wednesday", "Thursday"], 
            replace=True, 
            p=[0.1, 0.2, 0.3, 0.4]
            )
        
        # 判斷是否可以成功拿到書
        if date_to_return == "Monday" and location == "FL" or date_to_return == "Wednesday" and location == "CL":
            count += 1
    
    
    print(f"總共模擬了 {s} 次")
    print(f"成功拿到書的次數為: {count}")
    print(f"成功拿到書的機率為: {count/s:.4f}")
```

## 法二：在給定的條件下重新分配機率

```python
import numpy as np

# 模擬試驗次數
trials = 1000000

# 模擬借書者持書 4, 5, 6, 或 7 天的機率 但因為題目已經確定說不會在"第一天還在商圖"(因為第第四天的時候確定拿不到)
# 所以底下的機率要改寫
x = np.random.choice([4, 5, 6, 7], trials,
        replace=True, p=[0.05/0.95, 0.2/0.95, 0.3/0.95, 0.4/0.95])

# 記錄每個情況的次數
total_six_day_returns = 0
total_conditions_met = 0

# 遍歷每次試驗
for days in x:
  if days != 4 : #前面改寫機率後，就可以知道如果跑出來式第四天都一定是延遲的，所以只要是四都可以直接加
    # 模擬 50% 的機率是否將書歸還到其他圖書館，延遲 2 天
    is_foreign_library = np.random.choice([0, 1], p=[0.5, 0.5])

    # 增加額外 2 天延遲
    days += is_foreign_library * 2

    # 計算是否在第 6 天歸還
    if days == 6:
      total_six_day_returns += 1
  else :
    total_six_day_returns += 1

# 計算條件概率
probability = total_six_day_returns / trials


# 輸出結果
print(f"The probability of getting the book on the 6th day is approximately: {probability:.4f}")
```
